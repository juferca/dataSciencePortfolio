```python
# Import the necessary libraries.
import requests
import pandas as pd
import time

import pprint
```

```python
# My OMDb API key.
api_key = "your_api_key_here"
```

```python
# Load a list of movie titles from a local Excel (.xlsx) file.
titles_df = pd.read_excel("IMDb Top 100 Movies November 3, 2024.xlsx", sheet_name="Sheet2")
print("Loaded movie titles:", titles_df.head())
```

```python
titles_df
```

```python
# Define function to send a request to OMDb API for each title to fetch detailed movie data.
def fetch_movie_data(title):
    url = f"http://www.omdbapi.com/?t={title}&apikey={api_key}"

    try:
        response = requests.get(url, timeout=10)
        return response.json()

    except requests.exceptions.RequestException as e:
        print(f"Request error for {title}: {e}")
        return None
```

```python
# Loop over each title to fetch data from OMDb API and stores it in a list.
movie_data = []
for i, title in enumerate(titles_df["Movie Title"], start=1):
    print(f"Processing {i}: {title}")
    try:
        data = fetch_movie_data(title)
        if data and data.get("Response") == "True":
            movie_data.append(data)
        else:
            print(f"No data found for {title}")
    except Exception as e:
        print(f"Error with {title}: {e}")
    time.sleep(1)
```

```python
movie_data
```

```python
# Convert the list of movie data into a DataFrame.
movie_details_df = pd.DataFrame(movie_data)
print("Fetched movie details:", movie_details_df.head())
```

```python
movie_details_df.columns
```

```python
movie_details_df.dtypes
```

```python
movie_details_df.shape
```

```python
# Transformation step #1: Rename columns to more readable and standardized names.
movie_details_df = movie_details_df.rename(columns={
    'Title': 'title',
    'Year': 'year',
    'Rated': 'rating',
    'Released': 'release_date',
    'Runtime': 'runtime',
    'Genre': 'genre',
    'Director': 'director',
    'Writer': 'writer',
    'Actors': 'actors',
    'Plot': 'plot'
})
```

```python
movie_details_df.columns
```

```python
# Transformation step #2: Creates a separate 'rt_rating' column for Rotten Tomatoes ratings and removes the 'Ratings' column.
if 'Ratings' in movie_details_df.columns:
    movie_details_df['rt_rating'] = movie_details_df['Ratings'].apply(
        lambda ratings: next((rating['Value'] for rating in ratings if rating['Source'] == 'Rotten Tomatoes'), None)
    )
movie_details_df = movie_details_df.drop(columns=['Ratings'])
```

```python
movie_details_df.columns
```

```python
# Transformation step #3: Remove columns with all N/A values.
movie_details_df = movie_details_df.dropna(axis=1, how='all')
```

```python
# Transformation step #4: Remove rows with any N/A values.
movie_details_df = movie_details_df.dropna(axis=0, how='any')
```

```python
# Transformation step #5: Remove Language, Poster, Type, DVD, Production, Website and Response Columns to clean up data, focusing only on relevant 
# information.
movie_details_df = movie_details_df.drop(columns=['Language', 'Poster', 'Type', 'DVD', 'Production', 'Website', 'Response'], errors='ignore')
```

```python
movie_details_df.shape
```

```python
# Transformation step #6: Convert 'release_date' column to datetime format.
movie_details_df['release_date'] = pd.to_datetime(movie_details_df['release_date'], errors='coerce')
```

```python
# Transformation step #7: Extracts the numeric value from 'runtime' and converts it to integer (minutes).
movie_details_df['runtime'] = movie_details_df['runtime'].str.extract(r'(\d+)').astype(float)
```

```python
movie_details_df.dtypes
```

```python
movie_details_df.head()
```

```python
# Transformation step #8: Checks for duplicates based on 'title' and 'year' and remove them.
movie_details_df = movie_details_df.drop_duplicates(subset=['title', 'year'])
```

```python
movie_details_df.shape
```

```python
# Save cleaned dataset to a CSV file for future analysis.
movie_details_df.to_csv("cleaned_movie_dataset.csv", index=False)
print("Data saved to cleaned_movie_dataset.csv")
```
